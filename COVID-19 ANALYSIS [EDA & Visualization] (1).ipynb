```python
df_world = df[df.location == 'World']
df_world
```

### Finding The Date of Maximum Number of Deaths all over the World

```python
df_world[df_world['new_deaths'] == df_world['new_deaths'].max()]['date']
```

### creating a Summary table for the most recent date all over the world 

```python
df_world_recent = df_world[df_world['date'] == df_world.date.min()]
df_world_recent
```

```python
df_world_recent.groupby('date')[['total_cases','new_cases','new_deaths','total_deaths']].sum()
```

## Calculating the Percentage of Confirmed Cases Regarding the world POPULATION 

```python
df_world['population'].unique()
```

```python
df_world['total_cases'].max()
```

```python
df_world_ratio = df_world['total_cases'].max() / df_world['population'].max()
```

```python
df_world_ratio
```

### Visualizing World Data 

#### Confirmed Cases all over the world 

```python
df_world_overtime = df_world.groupby(['date'])[['total_cases','new_cases','new_deaths','total_deaths']].sum().reset_index().sort_values('date',ascending=True)
df_world_overtime
```

```python
# Using LINE_PLOT 

plt.figure(figsize=(20,10))
plt.plot(df_world_overtime.index, df_world_overtime['total_cases'])
plt.title('Evolution of Confirmed Cases of COVID_19', fontsize = 20)
plt.xlabel('Days',fontsize=20)
plt.ylabel('Confirmed Cases',fontsize=20)
```

### Total Deathes Cases Evolution Over Time 

```python
plt.figure(figsize=(20,10))
plt.plot(df_world_overtime.index, df_world_overtime['total_deaths'])
plt.title('Evolution of Deaths of COVID_19', fontsize = 20)
plt.xlabel('Days',fontsize=20)
plt.ylabel('Number of deaths',fontsize=20)
```

### New Cases all Over the World

```python
plt.figure(figsize=(20,10))
plt.plot(df_world_overtime.index, df_world_overtime['new_cases'])
plt.title('Evolution of new cases of COVID_19', fontsize = 20)
plt.xlabel('Days',fontsize=20)
plt.ylabel('New_cases',fontsize=20)
```

## Putting it all together 

```python
plt.figure(figsize=(20,10))
plt.plot(df_world_overtime.index, df_world_overtime['new_cases'], label = 'NEW')
plt.plot(df_world_overtime.index, df_world_overtime['total_cases'], label = 'Confirmed')
plt.plot(df_world_overtime.index, df_world_overtime['new_deaths'], label = 'New Deaths')
plt.plot(df_world_overtime.index, df_world_overtime['total_deaths'], label = 'Total_deaths')
plt.legend(loc=0)
```
